## Part A: NumPy Array Operations (30 points)

### Question 1: Revenue Simulation (10 points)

You run a small Nigerian market stall selling 4 products. Simulate 30 days of business:

**Requirements:**
1. Generate daily sales quantities for 4 products (use normal distribution)
   - Product A: mean=50, std=10
   - Product B: mean=30, std=5
   - Product C: mean=70, std=15
   - Product D: mean=40, std=8
2. Set random seed to your favorite number
3. Ensure no negative quantities (clip at 0)
4. Print the shape and first 5 days

**Deliverable:** Create a 30×4 NumPy array called `daily_sales`

```python
# Set random seed for reproducibility
np.random.seed(42)

# Define parameters for 4 products
products = ['Product A', 'Product B', 'Product C', 'Product D']
means = [50, 30, 70, 40]
stds = [10, 5, 15, 8]

# Generate daily sales quantities for 30 days (30 x 4 array)
daily_sales = np.random.normal(loc=means, scale=stds, size=(30, 4))

# Clip negative values to 0 (no negative sales)
daily_sales = np.clip(daily_sales, 0, None)

# Display results
print("Shape of daily_sales array:", daily_sales.shape)
print("\nFirst 5 days of sales:")
print(daily_sales[:5].astype(int))
print("\nProducts:", products)
```

### Question 2: Financial Analysis (10 points)

Using the `daily_sales` array from Q1:

**Calculate:**
1. Total sales for each product over 30 days
2. Average daily sales per product
3. The day with maximum total sales (across all products)
4. How many days had total sales above the 30-day average?
5. Standard deviation of sales for each product

**Deliverable:** Print all results with clear labels

```python
# 1. Total sales for each product over 30 days
total_sales_per_product = daily_sales.sum(axis=0)
print("1. Total sales for each product over 30 days:")
for i, product in enumerate(products):
    print(f"   {product}: {total_sales_per_product[i]:.2f} units")

# 2. Average daily sales per product
avg_daily_sales = daily_sales.mean(axis=0)
print("\n2. Average daily sales per product:")
for i, product in enumerate(products):
    print(f"   {product}: {avg_daily_sales[i]:.2f} units/day")

# 3. The day with maximum total sales (across all products)
daily_totals = daily_sales.sum(axis=1)
max_sales_day = np.argmax(daily_totals)
print(f"\n3. Day with maximum total sales: Day {max_sales_day + 1}")
print(f"   Total sales on that day: {daily_totals[max_sales_day]:.2f} units")

# 4. How many days had total sales above the 30-day average?
average_daily_total = daily_totals.mean()
days_above_average = np.sum(daily_totals > average_daily_total)
print(f"\n4. Days with total sales above 30-day average ({average_daily_total:.2f}): {days_above_average} days")

# 5. Standard deviation of sales for each product
sales_std = daily_sales.std(axis=0)
print("\n5. Standard deviation of sales for each product:")
for i, product in enumerate(products):
    print(f"   {product}: {sales_std[i]:.2f}")
```

### Question 3: Price & Revenue Calculation (10 points)

Your products have these prices: [200, 350, 150, 280] NGN

**Tasks:**
1. Calculate daily revenue (price × quantity for each product)
2. Calculate total revenue over 30 days
3. Apply a promotional discount:
   - Days 1-5: 15% off all products
   - Days 15-20: 10% off Product A and B only
   - Other days: no discount
4. Calculate total revenue with discounts
5. How much discount did you give away?

**Hint:** Use boolean indexing and broadcasting

```python
# Product prices in NGN
prices = np.array([200, 350, 150, 280])

# 1. Calculate daily revenue (price × quantity for each product)
daily_revenue = daily_sales * prices
print("1. Daily Revenue (first 5 days):")
print(daily_revenue[:5].astype(int))

# 2. Calculate total revenue over 30 days
total_revenue = daily_revenue.sum()
print(f"\n2. Total revenue over 30 days: ₦{total_revenue:,.2f}")

# 3. Apply promotional discounts
# Create a copy for discounted revenue
discounted_revenue = daily_revenue.copy()

# Days 1-5: 15% off all products (indices 0-4)
discounted_revenue[0:5, :] *= 0.85

# Days 15-20: 10% off Product A and B only (indices 14-19, columns 0-1)
discounted_revenue[14:20, 0:2] *= 0.90

# 4. Calculate total revenue with discounts
total_discounted_revenue = discounted_revenue.sum()
print(f"\n3 & 4. Total revenue with discounts: ₦{total_discounted_revenue:,.2f}")

# 5. How much discount was given away?
total_discount = total_revenue - total_discounted_revenue
print(f"\n5. Total discount given away: ₦{total_discount:,.2f}")

# Breakdown of discount by promotion
discount_days_1_5 = daily_revenue[0:5, :].sum() * 0.15
discount_days_15_20 = daily_revenue[14:20, 0:2].sum() * 0.10
print(f"   - Days 1-5 (15% off all): ₦{discount_days_1_5:,.2f}")
print(f"   - Days 15-20 (10% off A&B): ₦{discount_days_15_20:,.2f}")
```

**Dataset Description:**

This dataset simulates 500 Nigerian market transactions spanning 5 major markets (Bodija, Balogun, Ogbete, Wuse, Ariaria) across 5 states, covering 4 staple food products (Rice, Beans, Garri, Yam, Palm Oil) recorded daily from January 2024. 
Each record captures the vendor, product category, quantity sold, unit price in NGN, and total transaction amount, providing a realistic view of informal market trade patterns in Nigeria. 
The dataset also includes 35 missing values across the `quantity` and `unit_price` columns to simulate real-world data quality issues.

### Question 6: Aggregation & Grouping (15 points)

Using your dataset:

**Tasks:**
1. Group by a categorical column and calculate mean of a numeric column
2. Group by a categorical column and calculate multiple statistics (min, max, mean, sum)
3. Create a cross-tabulation or pivot table (if applicable)
4. Find the top 5 records by some criterion
5. Calculate at least one percentage or ratio

**Deliverable:** Perform 5 different aggregations/groupings

```python
print("=== 5 DIFFERENT AGGREGATIONS/GROUPINGS ===\n")

# 1. Group by categorical column and calculate mean of numeric column
print("1. Average quantity sold by product:")
avg_qty_by_product = df.groupby('product')['quantity'].mean().sort_values(ascending=False)
print(avg_qty_by_product)
print()

# 2. Group by categorical column and calculate multiple statistics (min, max, mean, sum)
print("2. Multiple statistics for total_amount by state:")
state_stats = df.groupby('state')['total_amount'].agg(['min', 'max', 'mean', 'sum']).round(2)
print(state_stats)
print()

# 3. Create a cross-tabulation (pivot table)
print("3. Cross-tabulation: Product vs State (total sales count):")
cross_tab = pd.crosstab(df['product'], df['state'], margins=True)
print(cross_tab)
print()

# 4. Find top 5 records by some criterion
print("4. Top 5 highest value transactions:")
top_5 = df.nlargest(5, 'total_amount')[['transaction_id', 'product', 'state', 'total_amount']]
print(top_5)
print()

# 5. Calculate percentage/ratio
print("5. Percentage of total revenue by product:")
revenue_by_product = df.groupby('product')['total_amount'].sum()
revenue_percentage = (revenue_by_product / revenue_by_product.sum() * 100).round(2)
revenue_df = pd.DataFrame({
    'Total Revenue': revenue_by_product,
    'Percentage': revenue_percentage
})
print(revenue_df)
```

### Question 7: Data Transformation (15 points)

Using your dataset:

**Tasks:**
1. Create a new column based on a calculation (e.g., total = price × quantity)
2. Create a categorical column based on numeric ranges (e.g., 'Low', 'Medium', 'High')
3. Sort data by 2 or more columns
4. Calculate a rolling or cumulative statistic (if time-series data)
5. Handle any missing values (if present) using appropriate methods

**Deliverable:** Show your transformed dataset with new columns

```python
# Work on a copy of the dataframe
df_transformed = df.copy()

print("=== DATA TRANSFORMATION ===\n")

# 1. Create a new column based on a calculation
print("1. Created 'revenue_per_unit' column (total_amount / quantity):")
df_transformed['revenue_per_unit'] = df_transformed['total_amount'] / df_transformed['quantity']
print(df_transformed[['product', 'quantity', 'total_amount', 'revenue_per_unit']].head())
print()

# 2. Create a categorical column based on numeric ranges
print("2. Created 'price_tier' column (Low, Medium, High, Premium):")
df_transformed['price_tier'] = pd.cut(
    df_transformed['unit_price'],
    bins=[0, 10000, 20000, 40000, float('inf')],
    labels=['Low', 'Medium', 'High', 'Premium']
)
print(df_transformed[['product', 'unit_price', 'price_tier']].head(10))
print()

# 3. Sort data by 2 or more columns
print("3. Sorted by state (ascending) and total_amount (descending):")
df_transformed = df_transformed.sort_values(['state', 'total_amount'], ascending=[True, False])
print(df_transformed[['transaction_id', 'state', 'product', 'total_amount']].head(10))
print()

# 4. Calculate a rolling or cumulative statistic
print("4. Calculated cumulative sum of total_amount by state:")
df_transformed = df_transformed.sort_values('date')  # Sort by date first
df_transformed['cumulative_revenue'] = df_transformed.groupby('state')['total_amount'].cumsum()
print(df_transformed[['date', 'state', 'total_amount', 'cumulative_revenue']].head(10))
print()

# 5. Handle missing values using appropriate methods
print("5. Handling missing values:")
print(f"   Missing before: quantity={df_transformed['quantity'].isnull().sum()}, unit_price={df_transformed['unit_price'].isnull().sum()}")

# Fill missing quantities with median for that product
df_transformed['quantity'] = df_transformed.groupby('product')['quantity'].transform(
    lambda x: x.fillna(x.median())
)

# Fill missing unit_prices with median for that product
df_transformed['unit_price'] = df_transformed.groupby('product')['unit_price'].transform(
    lambda x: x.fillna(x.median())
)

# Recalculate total_amount
df_transformed['total_amount'] = df_transformed['quantity'] * df_transformed['unit_price']

print(f"   Missing after: quantity={df_transformed['quantity'].isnull().sum()}, unit_price={df_transformed['unit_price'].isnull().sum()}")

print("\n=== TRANSFORMED DATASET SAMPLE ===")
print(df_transformed[['transaction_id', 'product', 'price_tier', 'cumulative_revenue']].head())
```

#### Question 8a: Which product generates the most revenue per region (state)?

```python
print("=== QUESTION 8a: Revenue by Product per State ===\n")

# Calculate total revenue by state and product
revenue_by_state_product = df.groupby(['state', 'product'])['total_amount'].sum().reset_index()

# Find top product in each state
top_product_by_state = revenue_by_state_product.loc[revenue_by_state_product.groupby('state')['total_amount'].idxmax()]

print("Top revenue-generating product in each state:")
for _, row in top_product_by_state.iterrows():
    print(f"  {row['state']}: {row['product']} (₦{row['total_amount']:,.2f})")

# Overall top product
total_by_product = df.groupby('product')['total_amount'].sum().sort_values(ascending=False)
print(f"\nOverall highest revenue product: {total_by_product.index[0]} (₦{total_by_product.iloc[0]:,.2f})")
```

**Interpretation:**

The analysis reveals which products are most profitable in each Nigerian state. This information helps vendors and suppliers understand regional preferences and optimize their distribution strategies. For example, if Rice dominates in Lagos but Palm Oil leads in Abuja, suppliers can tailor their inventory accordingly to maximize sales and reduce waste from unsold goods.

#### Question 8b: What is the price volatility (standard deviation) of each product across different markets?

```python
print("=== QUESTION 8b: Price Volatility by Product ===\n")

# Calculate price statistics by product across markets
price_volatility = df.groupby('product')['unit_price'].agg(['mean', 'std', 'min', 'max']).round(2)
price_volatility['coefficient_of_variation'] = (price_volatility['std'] / price_volatility['mean'] * 100).round(2)
price_volatility = price_volatility.sort_values('coefficient_of_variation', ascending=False)

print("Price volatility by product (sorted by coefficient of variation):")
print(price_volatility)

most_volatile = price_volatility.index[0]
least_volatile = price_volatility.index[-1]
print(f"\nMost volatile product: {most_volatile} (CV: {price_volatility.loc[most_volatile, 'coefficient_of_variation']}%)")
print(f"Least volatile product: {least_volatile} (CV: {price_volatility.loc[least_volatile, 'coefficient_of_variation']}%)")
```

**Interpretation:**

Price volatility analysis helps identify which products have unstable prices across markets. 
High volatility products (like seasonal crops) may require price hedging strategies or quick inventory turnover, 
while stable products (like staples) offer predictable margins. 
This insight is crucial for risk management in the Nigerian market, helping traders decide which products to stock based on their risk appetite.

#### Question 8c: Which vendors are the top performers by total sales volume and revenue?

```python
print("=== QUESTION 8c: Top Performing Vendors ===\n")

# Calculate metrics by vendor
vendor_performance = df.groupby('vendor').agg({
    'quantity': ['sum', 'mean'],
    'total_amount': ['sum', 'mean'],
    'transaction_id': 'count'
}).round(2)

# Flatten column names
vendor_performance.columns = ['_'.join(col).strip() for col in vendor_performance.columns]
vendor_performance = vendor_performance.rename(columns={'transaction_id_count': 'total_transactions'})

# Sort by total revenue
vendor_performance = vendor_performance.sort_values('total_amount_sum', ascending=False)

print("Vendor Performance (sorted by total revenue):")
print(vendor_performance)

# Calculate market share
total_revenue = df['total_amount'].sum()
vendor_performance['market_share_pct'] = (vendor_performance['total_amount_sum'] / total_revenue * 100).round(2)

print("\nVendor Market Share:")
for vendor in vendor_performance.index:
    share = vendor_performance.loc[vendor, 'market_share_pct']
    print(f"  {vendor}: {share}%")
```

**Interpretation:**

Vendor performance analysis reveals the key players in the market and their market share. This information is valuable for partnership decisions, competitive analysis, and identifying potential mentors for new vendors. Understanding which vendors generate the most revenue versus those with the highest transaction volume can also reveal different business models - high-volume low-margin versus low-volume high-margin strategies.
